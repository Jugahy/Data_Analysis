# 회귀계수

- 회귀 분석에서 **독립 변수**(데이터)와 **종속 변수**(타겟) 간의 관계를 수치적으로 나타내는 값
- 회귀계수는 독립 변수가 종속 변수에 얼마나 영향을 미치는지 즉 얼마나 변화를 일으키는지를 보여주는 수치


### **독립 변수**: 온도 데이터  
### **종속 변수**: 아이스크림 판매량 데이터  

온도와 아이스크림 판매량의 관계에서 **회귀계수**는 **온도가 1도 증가할 때 아이스크림 판매량이 얼마나 변하는지**를 수치적으로 나타냅니다. 따라서 온도가 올라가거나 내려갈 때 아이스크림 판매량이 어떻게 변화하는지를 예측하는 데 도움을 줍니다.

```python
import numpy as np

# 예제 데이터 (독립 변수: 온도, 종속 변수: 아이스크림 판매량)
X = np.array([1, 2, 3, 4, 5])  # 독립 변수 (온도)
y = np.array([3, 4, 2, 5, 6])  # 종속 변수 (아이스크림 판매량)
```

```python
print(X)
```

```python
print(y)
```

## 1. 선형 회귀식

## $$y = \beta_{0} + \beta_{1}x $$



## 2. 절편 & 기울기

### 회귀선을 그리기 위해서 절편과 기울기를 알아야 하며 **최소 제곱법**을 사용하여 구함




### 기울기 

```python
# 1. 독립 변수(X)와 종속 변수(y)의 평균 계산
mean_x = np.mean(X)
mean_y = np.mean(y)

# 2. 기울기 (β1) 계산
numerator = np.sum((X - mean_x) * (y - mean_y))  # ∑(xi - x̄)(yi - ȳ)
denominator = np.sum((X - mean_x) ** 2)          # ∑(xi - x̄)^2
beta_1 = numerator / denominator
```

### 절편
- 종속 변수의 평균 값에서 독립 변수의 평균 값을 이용해 구해진 기울기만큼을 뺀 값

평균 기준 독립 변수의 변화량이 적용되지 않았을 때 종속 변수의 기본 수준

```python
# 3. 절편 (β0) 계산
beta_0 = mean_y - beta_1 * mean_x
```

```python
print("기울기 (β1):", beta_1)
print("절편 (β0):", beta_0)
```

## 요약



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 예제 데이터 생성 (온도와 아이스크림 판매량)
np.random.seed(42)
temperature = np.random.randint(20, 40, 50)  # 20도에서 40도 사이의 임의 온도 데이터
sales = 5 * temperature + np.random.randint(-10, 10, 50)  # 아이스크림 판매량 = 5 * 온도 + noise

# 데이터프레임 생성
data = pd.DataFrame({'Temperature': temperature, 'IceCreamSales': sales})

# 독립 변수와 종속 변수 정의
X = data[['Temperature']]
y = data['IceCreamSales']

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# 회귀계수 및 절편 출력
print("회귀계수 (기울기):", model.coef_[0])
print("절편:", model.intercept_)

# 예측 수행 (온도에 따른 판매량 예측)
pred_sales = model.predict(X)


# 시각화
plt.scatter(temperature, sales, color='blue', label='Actual Sales')
plt.plot(temperature, pred_sales, color='red', label='Predicted Sales')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales')
plt.title('Ice Cream Sales vs Temperature')
plt.legend()
plt.show()
```
